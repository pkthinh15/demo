```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
from imutils import contours
import imutils
from utils import rotate_image,extract_digit,save_image,s
```

```python
im = cv2.imread("input/bang_diem.png")
s(im)
```

```python
rotated_image = rotate_image(im,-19.6)
s(rotated_image)
```

```python
im = rotated_image[330:910,260:1220]
s(im)
```

```python
gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize=3)
```

```python
minLineLength = 100
maxLineGap = 100
lines = cv2.HoughLinesP(edges,8,np.pi/180,100,minLineLength,maxLineGap)
print(lines.shape)
```

```python
lines = lines.reshape(lines.shape[0],lines.shape[2],lines.shape[1]).squeeze(-1)
lines.shape
```

```python
lines = sorted(lines,key=lambda x:x[1],reverse=False)
```

```python
dist = lines[3][1] - lines[2][1]
```

```python
for x1,y1,x2,y2 in lines[2:-2]:
    row_table = im[y1:y1+dist,:]
    s(row_table)
    extract_digit(row_table[:,:250])
    extract_digit(row_table[:,300:430])
    extract_digit(row_table[:,500:700])
    extract_digit(row_table[:,800:])
```
